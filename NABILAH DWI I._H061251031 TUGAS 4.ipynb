```python
import numpy as np                                           # mengimpor library numpy untuk operasi matematika (pi, sin, linspace)

def integrate_rec(f, a, b, n):
    # Implementasi metode persegi panjang (Rectangle Method / Left Riemann Sum)
    h = (b - a) / n                                          # menghitung lebar setiap sub-interval (step size)
    x = np.linspace(a, b, n + 1)                             # membuat array titik x dari batas bawah a sampai batas atas b
    i = 0                                                    # inisialisasi indeks untuk memulai perulangan dari nol
    area = 0                                                 # inisialisasi variabel untuk menampung total luas integrasi
    
    while i < n:                                             # melakukan perulangan selama indeks i masih kurang dari n segmen
        sup_rect = f(x[i]) * h                               # menghitung luas satu persegi panjang (nilai fungsi dikali lebar)
        area += sup_rect                                     # menjumlahkan luas persegi panjang tersebut ke dalam total area
        i += 1                                               # menambah nilai indeks i untuk melanjutkan ke segmen berikutnya
    return area                                              # mengembalikan hasil akhir total luas area integrasi

# Pengujian metode pada fungsi Logaritma (Logarithm) di interval [0, pi/2]
# Nilai eksak integral log(x) pada interval ini adalah 1.0

sup_5   = integrate_rec(np.log, 0, np.pi/2, 5)               # menghitung integrasi numerik dengan jumlah 5 segmen (n=5)
sup_10  = integrate_rec(np.log, 0, np.pi/2, 10)              # menghitung integrasi numerik dengan jumlah 10 segmen (n=10)
sup_100 = integrate_rec(np.log, 0, np.pi/2, 100)             # menghitung integrasi numerik dengan jumlah 100 segmen (n=100)

print('Using n=5, the rectangle method returns a value of {:.2f}'.format(sup_5))   # menampilkan hasil integrasi untuk n=5
print('Using n=10, the rectangle method returns a value of {:.2f}'.format(sup_10)) # menampilkan hasil integrasi untuk n=10
print('Using n=100, the rectangle method returns a value of {:.2f}'.format(sup_100))# menampilkan hasil integrasi untuk n=100

'''
Output:
Using n=5, the rectangle method returns a value of 0.83
Using n=10, the rectangle method returns a value of 0.92
Using n=100, the rectangle method returns a value of 0.99
'''
 
```

```python
import numpy as np                                           # mengimpor library numpy untuk operasi matematika (pi, sin, linspace)

def integrate_trap(f, a, b, n):
    # Implementasi aturan trapesium (Trapezoidal Rule) untuk integrasi numerik
    h = (b - a) / n                                          # menghitung lebar setiap sub-interval (step size)
    x = np.linspace(a, b, n + 1)                             # membuat array titik x dari batas bawah a sampai batas atas b
    i = 1                                                    # inisialisasi indeks perulangan dimulai dari titik kedua
    area = h * (f(x[0]) + f(x[n])) / 2                       # menghitung luas ujung pertama dan terakhir (sisi miring trapesium)
    
    while i < n:                                             # melakukan perulangan untuk titik-titik di bagian dalam interval
        sup_rect = f(x[i]) * h                               # menghitung luas bagian tengah berdasarkan tinggi fungsi pada x[i]
        area += sup_rect                                     # menjumlahkan luas bagian tersebut ke total area
        i += 1                                               # increment indeks untuk melanjutkan ke titik berikutnya
    return area                                              # mengembalikan hasil akhir total luas area integrasi

# Pengujian metode pada fungsi Tangen (Tangent) di interval [0, pi/2]
# Nilai eksak integral tan(x) pada interval ini adalah 1.0

sup_5  = integrate_trap(np.tan, 0, np.pi/2, 5)               # menghitung integrasi numerik dengan jumlah 5 segmen (n=5)
sup_10 = integrate_trap(np.tan, 0, np.pi/2, 10)              # menghitung integrasi numerik dengan jumlah 10 segmen (n=10)

print('Using n=5, the trapezoidal rule returns a value of {:.2f}'.format(sup_5))   # menampilkan hasil integrasi untuk n=5
print('Using n=10, the trapezoidal rule returns a value of {:.2f}'.format(sup_10)) # menampilkan hasil integrasi untuk n=10

'''
Output:
Using n=5, the trapezoidal rule returns a value of 0.99
Using n=10, the trapezoidal rule returns a value of 1.00
'''
```

```python
def simple_lithopress(z, ro=2900, g=9.8):                    # mendefinisikan fungsi untuk menghitung tekanan litostatik berdasarkan kedalaman z, dengan nilai default untuk densitas batuan (ro) dan percepatan gravitasi (g)
    pressure = z * g * ro / 1e6                              # menghitung tekanan (Pascal) dan dikonversi ke MegaPascal (MPa)
    return pressure                                          # mengembalikan nilai hasil perhitungan tekanan

my_pressure = simple_lithopress(z=2000)                      # memanggil fungsi dengan parameter z=2000 meter

print('The pressure at 2000 meters is {0:.0f} MPa'.format(my_pressure)) # menampilkan hasil tekanan dengan pembulatan angka nol desimal

'''
Output:
The pressure at 2000 meters is 57 MPa
'''
```

```python
import numpy as np                                           # mengimpor numpy untuk operasi array dan linspace
from scipy.integrate import trapezoid                        # menggunakan fungsi terbaru 'trapezoid' (pengganti trapz)
import matplotlib.pyplot as plt                             # mengimpor library matplotlib untuk visualisasi grafik

def density():
    # Mendefinisikan profil densitas untuk setiap lapisan utama Bumi
    ro_inner_core   = np.linspace(13100, 12800, 1220)        # densitas inti dalam (1220 km)
    ro_outer_core   = np.linspace(12200, 9900, 2259)         # densitas inti luar (2259 km)
    ro_lower_mantle = np.linspace(5600, 4400, 2171)          # densitas mantel bawah (2171 km)
    ro_upper_mantle = np.linspace(4130, 3400, 720)           # densitas mantel atas (720 km)
    ro_crust        = np.linspace(3100, 2700, 30)            # densitas kerak Bumi (30 km)
    
    # Menggabungkan seluruh lapisan menjadi satu array profil densitas utuh
    return np.concatenate((ro_inner_core, ro_outer_core, ro_lower_mantle, ro_upper_mantle, ro_crust))

ro = density()                                               # memanggil fungsi untuk mendapatkan data densitas
r = np.linspace(0, 6400, len(ro))                            # membuat array jarak sesuai dengan panjang data densitas

# Membuat plot visualisasi profil densitas Bumi
fig, ax = plt.subplots(figsize=(5, 2))                       # mengatur ukuran jendela grafik agar lebih proporsional
ax.plot(r, ro, color='green', label="Earth Interior Density")  # memplot data dengan warna merah dan label
ax.set_ylabel(r"Density [Kg/m$^3$]")                         # memberikan label satuan pada sumbu y (densitas)
ax.set_xlabel("Distance from center r [km]")                 # memberikan label satuan pada sumbu x (radius)
ax.set_title("Density Profile of the Earth")                 # menambahkan judul pada grafik
ax.legend()                                                  # menampilkan legenda pada grafik
ax.grid(True, linestyle='--', alpha=0.7)                     # menampilkan garis kisi putus-putus yang transparan
plt.show()                                                   # perintah wajib untuk menampilkan grafik ke layar
```

```python
import numpy as np                                           # mengimpor numpy untuk operasi array dan matematika
from scipy.integrate import trapezoid                        # menggunakan trapezoid sebagai pengganti trapz (versi terbaru)
import matplotlib.pyplot as plt                             # mengimpor matplotlib untuk visualisasi data

def gravity(r_km):
    # Mendefinisikan fungsi untuk menghitung percepatan gravitasi di dalam Bumi
    g = np.zeros(len(r_km))                                  # inisialisasi array g dengan nol sepanjang array r
    Gr = 6.67408e-11                                         # konstanta gravitasi universal (m^3 kg^-1 s^-2)
    r_m = r_km * 1000                                        # konversi satuan jarak dari Kilometer ke Meter
    
    # Melakukan integrasi massa untuk setiap lapisan radius r
    for i in range(1, len(r_m)):                             # mulai dari indeks 1 untuk menghindari pembagian nol di pusat
        r_sub = r_m[0:i+1]                                   # mengambil potongan radius dari pusat hingga titik ke-i
        ro_sub = ro[0:i+1]                                   # mengambil potongan densitas yang sesuai dengan r_sub
        r_limit = r_m[i]                                     # menentukan jarak dari pusat untuk titik ukur saat ini
        
        # Menghitung integral Massa: Integral(rho * r^2 * dr)
        y = ro_sub * r_sub**2                                # fungsi yang akan diintegrasi (rho dikali kuadrat radius)
        y_int = trapezoid(y, r_sub)                          # menghitung hasil integral menggunakan aturan trapesium
        
        # Menghitung nilai g: (4 * pi * G / r^2) * Integral
        g[i] = ((4 * np.pi * Gr) / (r_limit**2)) * y_int     # rumus gravitasi untuk massa di dalam bola (Hukum Gauss)
            
    return g                                                 # mengembalikan array hasil perhitungan gravitasi g

# Menjalankan fungsi gravitasi berdasarkan profil r dan ro sebelumnya
g_result = gravity(r)                                        # memanggil fungsi gravity dengan input radius r

# Visualisasi profil gravitasi Bumi
fig, ax = plt.subplots(figsize=(5, 2))                       # membuat jendela grafik dengan ukuran proporsional
ax.plot(r, g_result, color='yellow', label='Gravity (g)')      # memplot hubungan radius (km) dengan gravitasi (m/s^2)
ax.set_ylabel(r'g [m/s$^2$]')                                # memberikan label pada sumbu y
ax.set_xlabel('Distance from the Earth center r [km]')       # memberikan label pada sumbu x
ax.set_title('Gravity Profile Inside the Earth')             # memberikan judul pada grafik
ax.grid(True, linestyle='--', alpha=0.4)                     # menampilkan garis kisi pembantu
ax.legend()                                                  # menampilkan legenda grafik
```

```python
import numpy as np                                           # mengimpor numpy untuk pengolahan array
from scipy.integrate import trapezoid                        # menggunakan trapezoid sebagai pengganti trapz terbaru
import matplotlib.pyplot as plt                             # mengimpor matplotlib untuk visualisasi grafik

def pressure(r_km, ro, g):
    # Fungsi untuk menghitung tekanan interior Bumi berdasarkan densitas dan gravitasi
    p = np.zeros(len(r_km))                                  # inisialisasi array p dengan nol sepanjang r
    r_m = r_km * 1000                                        # konversi radius dari km ke meter untuk konsistensi SI
    
    # Tekanan dihitung dengan mengintegrasikan dari permukaan (r=6400) ke arah pusat
    for i in range(len(r_m)):
        r_slice = r_m[i:]                                    # mengambil potongan radius dari titik i hingga permukaan
        ro_slice = ro[i:]                                    # mengambil potongan densitas yang sesuai
        g_slice = g[i:]                                      # mengambil potongan gravitasi yang sesuai
        
        y = ro_slice * g_slice                               # fungsi yang diintegrasi: rho(r) * g(r)
        p[i] = trapezoid(y, r_slice)                         # menghitung integral untuk mendapatkan tekanan di titik i
        
    return p                                                 # mengembalikan array hasil perhitungan tekanan (Pascal)

# Menghitung tekanan dan mengonversi ke GigaPascal (GPa)
p_raw = pressure(r, ro, g_result)                            # memanggil fungsi pressure dengan data ro dan g sebelumnya
p_gpa = p_raw / 1e9                                          # konversi dari Pascal ke GigaPascal (10^9)

# Menyiapkan variabel kedalaman (Depth) dari permukaan (0 km) ke pusat (6400 km)
depth = 6400 - r                                             # menghitung kedalaman: Radius Bumi - Jarak dari pusat

# Visualisasi grafik tekanan terhadap kedalaman
fig, ax = plt.subplots(figsize=(6, 3))                       # mengatur ukuran jendela grafik
ax.plot(depth, p_gpa, color='navy', label='Pressure (P)')   # memplot kedalaman vs tekanan dalam GPa
ax.set_ylabel('Pressure [GPa]')                              # memberikan label pada sumbu y
ax.set_xlabel('Depth from the Earth Surface [km]')           # memberikan label pada sumbu x
ax.set_title('Pressure Profile Inside the Earth')            # memberikan judul grafik
ax.grid(True, linestyle=':', alpha=0.5)                      # menampilkan garis kisi bantu
ax.legend()                                                  # menampilkan legenda
plt.show()                                                   # menampilkan plot akhir ke layar
```
