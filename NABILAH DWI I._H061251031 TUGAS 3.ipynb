```python
import pandas as pd #untuk membaca file excel
from scipy.stats.mstats import gmean, hmean #untuk menghitung geometric mean dan harmonic mean
import matplotlib.pyplot as plt #untuk membuat grafik histogram

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces') #Menggunakan pandas untuk membaca file Excel

#Data dari sheet Supp_traces dimasukkan ke dalam variabel my_dataset

a_mean = my_dataset.Ho .mean() #Menghitung rata-rata aritmatika dari kolom Ho
g_mean = gmean(my_dataset['Ho']) #Menghitung rata-rata geometri menggunakan fungsi dari SciPy
h_mean = hmean(my_dataset['Ho']) #Menghitung rata-rata harmonik

print ('-------')#garis pemisah di output
print ('arithmetic mean')#menampilkan teks 'arithmetic mean' di output
print ("{0:.1f} [ppm]".format(a_mean))#menampilkan nilai rata-rata aritmatika dengan format satu angka di belakang koma dan satuan ppm
print ('-------')#garis pemisah di output

print ('geometric mean')#menampilkan teks 'geometric mean' di output
print ("{0:.1f} [ppm]".format(g_mean))#menampilkan nilai rata-rata geometri dengan format satu angka di belakang koma dan satuan ppm
print ('-------')#garis pemisah di output

print ('harmonic mean')#menampilkan teks 'harmonic mean' di output
print ("{0:.1f} [ppm]".format(h_mean))#menampilkan nilai rata-rata harmonik dengan format satu angka di belakang koma dan satuan ppm
print ('-------')#garis pemisah di output

fig, ax = plt.subplots() #Membuat sebuah figure dan axes untuk plotting menggunakan Matplotlib
ax.hist(my_dataset.Ho, bins='auto', density=True, edgecolor='k', label='Measurements Hist', alpha=0.8) #Membuat histogram dari data Ho dengan jumlah bin otomatis, menormalkan histogram, memberikan warna tepi pada batang histogram, menambahkan label untuk legenda, dan mengatur transparansi batang histogram
ax.axvline(a_mean, color='purple', label='Arithmetic mean', linewidth=2) #Menambahkan garis vertikal pada posisi rata-rata aritmatika dengan warna ungu, label untuk legenda, dan ketebalan garis 2
ax.axvline(g_mean, color='Blue',  label='Geometric mean', linewidth=2) #Menambahkan garis vertikal pada posisi rata-rata geometri dengan warna biru, label untuk legenda, dan ketebalan garis 2
ax.axvline(h_mean, color='pink',  label='Harmonic mean', linewidth=2) #Menambahkan garis vertikal pada posisi rata-rata geometri dengan warna biru, label untuk legenda, dan ketebalan garis 2
ax.set_xlabel('Ho [ppm]') #Menetapkan label untuk sumbu x sebagai 'Ho [ppm]'
ax.set_ylabel('Probability density') #Menetapkan label untuk sumbu y sebagai 'Probability density'
ax.legend() #Menampilkan legenda pada plot untuk membedakan antara histogram dan garis rata-rata

'''
Output:
-------
arithmetic mean
365.4 [ppm]
-------
geometric mean
348.6 [ppm]
-------
harmonic mean
333.8 [ppm]
-------
'''
```

```python
import pandas as pd #untuk membaca file excel
import numpy as np #untuk menghitung persentil
import matplotlib.pyplot as plt #untuk membuat grafik histogram

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces') #Menggunakan pandas untuk membaca file Excel

q1 = np.percentile(my_dataset.Hf, 25, method = 'midpoint') #Menghitung kuartil pertama (Q1) dan kuartil ketiga (Q3) dari data Hf menggunakan fungsi np.percentile dengan metode 'midpoint' untuk menentukan nilai kuartil
q3 = np.percentile(my_dataset.Hf, 75, method = 'midpoint') #Menghitung kuartil pertama (Q1) dan kuartil ketiga (Q3) dari data Hf menggunakan fungsi np.percentile dengan metode 'midpoint' untuk menentukan nilai kuartil
  
iqr = q3 - q1 # Interquaritle range (IQR)  

print ('-------') #garis pemisah di output
print ('Interquaritle range (IQR): {0:.0f} [ppm]'.format(iqr)) #Menampilkan nilai interkuartil range (IQR) dengan format tanpa angka di belakang koma dan satuan ppm
print ('-------') #garis pemisah di output

fig, ax = plt.subplots() #Membuat sebuah figure dan axes untuk plotting menggunakan Matplotlib
ax.hist(my_dataset.Hf, bins='auto', density=True, edgecolor='k', label='Measurements Hist') #Membuat histogram dari data Hf dengan jumlah bin otomatis, menormalkan histogram, memberikan warna tepi pada batang histogram, dan menambahkan label untuk legenda
ax.axvline(q1, color='khaki', label='Q1', linewidth=2) #Menambahkan garis vertikal pada posisi kuartil pertama (Q1) dengan warna khaki, label untuk legenda, dan ketebalan garis 2
ax.axvline(q3, color='cyan', label='Q3', linewidth=2) #Menambahkan garis vertikal pada posisi kuartil ketiga (Q3) dengan warna cyan, label untuk legenda, dan ketebalan garis 2
ax.axvspan(q1, q3, alpha=0.1, color='orange', label='Interquaritle range (IQR)') #Menambahkan area vertikal yang diarsir antara kuartil pertama (Q1) dan kuartil ketiga (Q3) dengan transparansi 0.1, warna oranye, dan label untuk legenda yang menunjukkan 'Interquaritle range (IQR)'
ax.set_xlabel('Hf [ppm]') #Menetapkan label untuk sumbu x sebagai 'Hf [ppm]'
ax.set_ylabel('Probability density') #Menetapkan label untuk sumbu y sebagai 'Probability density'
ax.legend() #Menampilkan legenda pada plot untuk membedakan antara histogram, garis kuartil pertama (Q1), garis kuartil ketiga (Q3), dan area interkuartil range (IQR)

'''
Output:
-------
Interquaritle range (IQR): 164 [ppm]
-------
'''
```

```python
import pandas as pd #untuk membaca file excel
import matplotlib.pyplot as plt #untuk membuat grafik boxplot

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces') #Menggunakan pandas untuk membaca file Excel

fig, ax = plt.subplots() #Membuat sebuah figure dan axes untuk plotting menggunakan Matplotlib
my_flierprops = dict(markerfacecolor="#34c1ae", markeredgecolor="#d423c69d", marker='s') #Mengatur properti untuk outliers (flier) pada boxplot dengan warna marker hijau toska, warna tepi marker merah muda transparan, dan bentuk marker kotak
my_medianprops = dict(color="#7d77d1", linewidth = 2) #Mengatur properti untuk garis median pada boxplot dengan warna ungu dan ketebalan garis 2
my_boxprops = dict(facecolor="#afc84c", edgecolor='#24305e') #Mengatur properti untuk kotak pada boxplot dengan warna isi hijau kekuningan dan warna tepi biru tua
ax.boxplot(my_dataset.U, patch_artist = True, notch=True, flierprops=my_flierprops, medianprops=my_medianprops, boxprops=my_boxprops) #Membuat boxplot dari data U dengan menggunakan properti yang telah diatur untuk outliers, median, dan kotak, serta mengaktifkan notch pada boxplot
ax.set_ylabel('U [ppm]') #Menetapkan label untuk sumbu y sebagai 'U [ppm]'
ax.set_xticks([1]) #Menetapkan posisi tick pada sumbu x di posisi 1
ax.set_xticklabels(['all Epochs']) #Menetapkan label untuk tick pada sumbu x sebagai 'all Epochs'
plt.show() #Menampilkan plot boxplot yang telah dibuat  
```
